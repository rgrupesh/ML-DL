```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras import layers, losses
from tensorflow.keras.layers import Dropout
```

```python
PATH= "http://storage.googleapis.com/download.tensorflow.org/data/ecg.csv"
df = pd.read_csv(PATH,header=None)
df.head()
```

```python
label=df[140]
features=df.drop(labels=140,axis=1)
```

```python
features.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, label, train_size=0.8, random_state=42)
```

```python
train_index = y_train[y_train == 1].index
```

```python
train_index
```

```python
train_data = x_train.loc[train_index]
```

```python
train_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler(feature_range=(0, 1))
x_train_scaled = min_max_scaler.fit_transform(train_data.copy())
x_test_scaled = min_max_scaler.transform(x_test.copy())
```

```python
class AnomalyDetector(Model):
  def __init__(self):
    super(AnomalyDetector, self).__init__()
    self.encoder = tf.keras.Sequential([
      layers.Dense(32, activation="relu"),
      Dropout(0.1),
      layers.Dense(16, activation="relu"),
      Dropout(0.1),
      layers.Dense(8, activation="relu")])
    
    self.decoder = tf.keras.Sequential([
      layers.Dense(16, activation="relu"),
      Dropout(0.1),
      layers.Dense(32, activation="relu"),
      Dropout(0.1),
      layers.Dense(140, activation="sigmoid")])
    
  def call(self, x):
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded

autoencoder = AnomalyDetector()
```

```python
autoencoder.compile(optimizer='adam', loss='mae',metrics=['acc'])
```

```python
history = autoencoder.fit(
    x_train_scaled,
    x_train_scaled,
    epochs=20,
    batch_size=512,
    validation_data=(x_test_scaled, x_test_scaled),
    verbose=1
)
```
